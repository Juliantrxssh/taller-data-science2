Escribe el código para convertir el siguiente número que representa el ID de empleado "1001" a un número entero.

```python
id_empleado_str = "1001"

id_empleado_int = int(id_empleado_str)

print(f"Valor: {id_empleado_int}")
print(f"Tipo de dato: {type(id_empleado_int)}")
```

Escribe el código para convertir este número que representa el valor financiero "1234.56" a un número de punto flotante.

```python
valor_str = "1234.56"

valor_float = float(valor_str)

print(f"Valor: {valor_float}")
print(f"Tipo de dato: {type(valor_float)}")
```
